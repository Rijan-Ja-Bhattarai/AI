```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Dataset
dataset = pd.read_csv("C:/Users/ACER/Documents/GitHub/AI/Pandas Practice Problems/data.csv")
```

# Clean the Dataset

```python
dataset["Date"] = dataset["Date"].replace('20201226', '2020/12/26')
dataset["Date"] = dataset["Date"].fillna(method = "ffill")
```

```python
dataset["Date"] = pd.to_datetime(dataset["Date"], format = 'mixed')
```

```python
dataset["Year"] = dataset["Date"].dt.year
dataset["Month"] = dataset["Date"].dt.month
dataset["Day"] = dataset["Date"].dt.day
dataset.drop("Date", axis = 1)
```

```python
dataset["Calories"] = dataset["Calories"].fillna(dataset["Calories"].mean())
```

# Data Visualization

```python
day = dataset["Day"]
calories = dataset["Calories"]

plt.title("Calories Intake In a Month")
plt.bar(day, calories, color = 'g')

plt.grid(True)
plt.xlabel("Day")
plt.ylabel("Calories")
plt.tight_layout()
plt.show()
```

# Sigmoid Function

```python
def sigmoid(x, w, b):
    e = math.e
    z = np.dot(x, w) + b

    sigmoid = 1 / (1 + (e) ** -z)

    return sigmoid
```

```python
sig = sigmoid(x, w_init, b_init)
print(f"sigmoid: {sig}")
```

```python
z = np.arange(-10, 10).reshape(-1, 1)
plt.plot(z, sig)
```

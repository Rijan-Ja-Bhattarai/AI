# Create a random tensor of (7, 7)

```python
import torch
```

```python
size_seven_by_seven_tensor = torch.rand(size = (1, 7, 7))
size_seven_by_seven_tensor
```

# Perform Matrix Multiplication Between 2 Tensors

```python
tensor_A = torch.rand(size = (1, 1, 7))
tensor_A
```

```python
# Peform matmul
mul_tensor = torch.matmul(tensor_A, size_seven_by_seven_tensor)
mul_tensor
```

# Create a Random Seed and do the matmul again

```python
torch.manual_seed(777)
size_seven_by_seven_tensor = torch.rand(size = (1, 7, 7))

torch.manual_seed(777)
tensor_A = torch.rand(size = (1, 1, 7))

print(f"Tensor A: \n{size_seven_by_seven_tensor}")
print(f"Tensor B: \n{tensor_A}")

print(f"Shape of Tensor A: {size_seven_by_seven_tensor.shape}")
print(f"Shape of Tensor B: {tensor_A.shape}")


print(f"{size_seven_by_seven_tensor == tensor_A}")

mul_tensor = torch.matmul(tensor_A, size_seven_by_seven_tensor)
print(f"Multiplied Tensors: \n{mul_tensor}")
print(f"Shape of Multiplied Tensor: {mul_tensor.shape}")
```

# Set the random seed on the GPU

```python
torch.cuda.manual_seed(1234)
tensor_A = torch.rand(3, 4, device="cuda")
tensor_A
```

# Send Two Tensors From CPU to The GPU

```python
torch.cuda.manual_seed(1234)
tensor_A = torch.rand(2, 3)

torch.cuda.manual_seed(1234)
tensor_B = torch.rand(2, 3)

tensor_A.device, tensor_B.device
```

```python
tensor_A_gpu = tensor_A.to("cuda")
tensor_B_gpu = tensor_B.to("cuda")

tensor_A_gpu.device, tensor_B_gpu.device
```

# Matmul of Two Tensors in the GPU

```python
gpu_mat_mul = tensor_A_gpu @ tensor_B_gpu.T
print(f"Matrix Multiplication on GPU: {gpu_mat_mul}")
```

# Finding the Max and Min of Matmul Matrix

```python
max_val = gpu_mat_mul.max()
min_val = gpu_mat_mul.min()
max_val_index = gpu_mat_mul.argmax()
min_val_index = gpu_mat_mul.argmin()

print(f"Max Value: {max_val}")
print(f"Min Value: {min_val}")
print(f"Max Value Index: {max_val_index}")
print(f"Min Value Index: {min_val_index}")
```

# Creating a tensor and removing 1 dimensions from it

# Fitness Tracker

```python
import numpy as np
import matplotlib.pyplot as plt


# Parameters
one_month = 28
min_steps = 4000
max_steps = 12000
min_calories = 1800
max_calories = 3200
min_sleep = 5.0
max_sleep = 9.0

# Create Fitness Data for 4 weeks

steps = np.random.randint(min_steps, max_steps, size = one_month)
calories = np.random.randint(min_calories, max_calories, size = one_month)
sleep = np.random.uniform(min_sleep, max_sleep, size = one_month).round(1)

# Ideal Conditions
goal_steps = np.full(one_month, 10000)
goal_calories = np.full(one_month, 2500)
goal_sleep = np.full(one_month, 28)

# Normalize Data
normalized_steps = (steps - steps.min()) / (steps.max() - steps.min())
normalized_calories = (calories - calories.min()) / (calories.max() - calories.min())
normalized_sleep = (sleep - sleep.min()) / (sleep.max() - sleep.min())

# Plot the Data
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(normalized_steps, color='red')
plt.title('Normalized Steps')

plt.subplot(3, 1, 2)
plt.plot(normalized_calories, color='green')
plt.title('Normalized Calories')

plt.subplot(3, 1, 3)
plt.plot(normalized_sleep, color='blue')
plt.title('Normalized Sleep')

plt.tight_layout()
plt.show()
```
